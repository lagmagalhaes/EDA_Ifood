```python
# Importar Bibliotecas

# Importar bibioteca para manipulação de dados
import pandas as pd

# Importar biblioteca para operações matematicas
import numpy as np

# Importando biblotecas para auxiliar na visualização dos dados
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados e ler o arquivo.csv

df = pd.read_csv("/content/drive/MyDrive/mkt_data.csv")
```

Suponha que estamos a tentar atrair mais clientes. Na análise de dados anterior, vimos que os clientes sem filhos gastam mais que os clientes com filhos. A equipa de marketing pensou em fazer campanhas direcionadas ao grupo de pessoas com filho (independente da quantidade), porém precisamos comprovar se de fato as médias de gastos dos grupos é diferente. Portanto:

**Agrupar todas os clientes que tem filhos em um grupo chamado "com filhos" e fazer uma comparação da média de gastos de clientes sem filhos e clientes com filhos.**


```python
# Criando coluna de despesas dos que tem filhos

df['expenses_com_kids'] = df[df['kids']!=0]['expenses']

# Criando coluna de despesas dos que não tem filhos

df['expenses_sem_kids'] = df[df['kids']== 0]['expenses']

# Mostrar o Arquivo
df
```

Aqui não conhecemos o desvio padrão populacional, apenas conseguimos o desvio padrão e média das amostras. Vamos verificar a quantidade de dados de cada uma das amostras para verificar se podemos aplicar o Teorema do Limite Central.

```python
# Contar o número de valores não nulos na coluna "expenses_com_filhos".
df['expenses_com_kids'].count()
```

```python
# Contar o número de valores não nulos na coluna "expenses_sem_filhos".
df['expenses_sem_kids'].count()
```

Temos um grande número de amostras, pelo que o ***Teorema do Limite Central*** assegura que a distribuição das médias das amostras se aproxima de uma distribuição normal. Assim, vamos proceder à aplicação do ***Teste T***.

Ho: as médias de gastos são iguais.

Ha: as médias de gastos são diferentes.

Vamos usar 95% de confiança

```python
from scipy import stats
import numpy as np

# Suponha que df é o DataFrame com as colunas expenses_com_kids e expenses_sem_kids

# Removendo valores NaN e obtendo as amostras

expenses_com_kids = df['expenses_com_kids'].dropna()
expenses_sem_kids = df['expenses_sem_kids'].dropna()

# Calculando o teste t
t_stat, p_value = stats.ttest_ind(expenses_com_kids, expenses_sem_kids)

# Exibindo os resultados
print("Estatística t:", t_stat)
print("Valor-p:", p_value)
```


O p-value é muito menor que 0.05, logo rejeitamos a hipotese nula e podemos afirmar ao nível de 95% de confiança que houve diferença na média de gastos entre os grupos que tem filhos e os que não tem filhos.

```python
#calcular a média para cada grupo
print("Médias expenses com filhos:", expenses_com_kids.mean().round(2))
print("Médias expenses sem filhos:", expenses_sem_kids.mean().round(2))
```

Conseguimos ver claramente que a média de gastos para os clientes com filhos é claramente menor dque os clientes sem filhos.



**Agora, vamos comparar os grupos de pessoas com filhos. O time de marketing nos perdiu para verificar se há diferenças de gastos em quem tem 1, 2, 3 ou mais filhos para que eles possam afunilar mais ainda a campanha. Para isso, vamos analisar os dados:**

```python
# Verificar o tamanho de cada amostra

df['kids'].value_counts()
```

```python
df.groupby(["kids"]).expenses.mean().round(2)
```

Podemos constatar que o nosso objetivo é analisar se o número de filhos tem impacto na média de despesas e comparar os grupos. Para isso, vamos utilizar a ANOVA 1 fator. As permissas para a ANOVA são: Premissas:

Amostras aleatorias;

Grupos independentes;
A variáveis dependente é contínua
A variavel independente é categorica os dados devem seguir uma distribuição normal, ou cada grupo deve ter mais de 15 ou 20 observações
Os grupos devem ter variaçoes aproximadamente iguais.

Ho: Não há diferenças na média de gastos entre os grupos.

Ha: Nem todas as médias são iguais.

**Antes de realizar o teste ANOVA, é necessário verificar se todas as variâncias são iguais:**

```python
# Eliminar os valores nulos para fazer os testes

com_um_filho = df[df['kids'] == 1]['expenses'].dropna()
com_dois_filhos = df[df['kids'] == 2]['expenses'].dropna()
com_tres_filhos = df[df['kids'] == 3]['expenses'].dropna()
```

```python
# Vamos usar o teste de Barlett para testar se as variâncias são iguais ou diferentes

import scipy.stats as stats

stats.bartlett(com_um_filho, com_dois_filhos, com_tres_filhos)
```

O p_value é menor que 0.05, logo rejeitamos a hipotese nula e aceitamos que pelo menos uma das amostras possui variancia diferente. Portanto, poderiamos usar a **ANOVA de Welch**

```python
!pip install pingouin
```

```python
import pingouin as pg

com_filhos = df[df['kids'] != 0]

pg.welch_anova(dv='expenses',between='kids',data=com_filhos)
```

O p_valor é menor do que alpha (0.05), portanto, rejeitamos H0 e concluimos que a 95% de confiança que as médias de gastos entre os grupos não são iguais.

Agora, vamos usar um teste **Post Hoc (Games-Howell)** para descobrir qual dos pares não são iguais. Lembrando que esses dados violam a aproximação de dados para normalidade e as amostras não tem variancias iguais. Neste caso não podemos usar Tukey.

```python
# Utilizar o teste games Howell

pg.pairwise_gameshowell(dv='expenses',between='kids',data=com_filhos)
```

Apenas quando p-valor for maior do que 5e-02 aceitamos H0 (médias iguais). Logo, todas as médias são consideradas diferentes, exceto a 2-3.

Verificamos que os grupos dos clientes com 1 filho e 2 filhos têm médias diferentes, uma vez que o p_valor é inferior a 0,05, permitindo-nos rejeitar a hipótese nula. O mesmo acontece com o grupo dos clientes com 1 filho e 3 filhos. No entanto, para o grupo de pessoas com 2 filhos e 3 filhos, o p_valor é superior a 0,05, pelo que não podemos rejeitar a hipótese nula.

Sendo assim, os gastos médios dos grupos com 2 e 3 filhos são estatisticamente iguais. No entanto, verificam-se diferenças significativas nas médias dos restantes grupos. Consequentemente, a equipa de marketing pode direcionar os seus esforços para os grupos familiares com 2 ou 3 filhos
